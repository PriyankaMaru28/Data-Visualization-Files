QUESTION NO 1

The following violin plot shows the height distributions of athletes in the Gymnastics, Cycling and
Basketball sports at the Olympic games between 2000 to 2016 (inclusive). The line inside each
violin shows the location of the median value.Recreate the above violin plot using Matplotlib. The three violins should be purple, green and blue
in colour, including their filled colour and the colour of the median lines (Hint: consider the return
values of the violin plot method). Include the x-axis and y-axis labels, as well as the plot title.Since some athletes may compete in multiple events or multiple Olympics, make sure to include
only one instance of an athlete (no duplicates).

```python
df_sports = df[(df.Year >= 2000) & (df.Year <= 2016)]
df_sports 
```

```python
df_sports = df_sports.drop_duplicates(['ID']).dropna(subset=['Height'])
```

```python
df_sport_bas = df_sports[(df_sports.Sport == "Basketball") ]
df_sport_cyc = df_sports[(df_sports.Sport == "Cycling") ]
df_sport_gym = df_sports[(df_sports.Sport == "Gymnastics") ]
```

```python
df_list = list([df_sport_gym['Height'],df_sport_cyc['Height'],df_sport_bas['Height']])
df_list
```

```python
print('QUESTION 1')
fig, ax = plt.subplots(figsize=(9,6))

# build a violin plot
vp_plot = ax.violinplot(df_list,positions=None, vert=False, widths=0.9,showmedians=True, showmeans=False, showextrema=False)

# add title and axis labels
ax.set_title('Height Distributions (2000 - 2016 Olympics)')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Sport')

yticklabels = ['Gymnastics','Cycling','Basketball']
ax.set_yticks([1,2,3])
ax.set_yticklabels(yticklabels)

ax.set_xticks([140,160,180,200,220])
ax.set_xticklabels([140,160,180,200,220])

ax.set_xlim(130,230)
colors=['purple','green','blue']

len_colors = len(colors)
j = 0
for ac in vp_plot['bodies']:
    ac.set_facecolor(colors[j])
    j += 1
    if j == len_colors:
        j = 0
vp_plot['cmedians'].set_color('black')

plt.show()
```

QUESTION NO 2

This line plot shows the average height of male (blue) and female (red) Olympians over the history
of the Summer Olympics. The shaded region represents the standard deviation about the average.Recreate the above line plot using Matplotlib. Include the axes labels, plot title, and legend in the
lower left corner. The average male heights should be blue and average female heights red. Each
athlete should only be represented once in a given year.
Each line should include shading that has width equal to twice the standard deviation, that is, the
top of the shaded region is the average value plus the standard deviation, and the bottom of the
shaded region is the average value minus the standard deviation (hint: try fill_between()).
The shading uses a transparency value of 0.1.

```python
df_sum = df[df.Season == 'Summer'].sort_values(by='Year')
df_gen = df_sum[(df_sum.Year >= 1900) & (df_sum.Year <= 2020)].drop_duplicates(['ID']).dropna(subset=['Height'])
print('....height..',df_gen[df_gen.Height >200])
```

```python
#Plotting
df_gen_male = df_gen[(df_gen.Sex == 'M') ].groupby(['Year']).mean().reset_index()
df_gen_female = df_gen[(df_gen.Sex == 'F')].groupby(['Year']).mean().reset_index()
```

```python
#Shadding
df_gen_male_1 = df_gen[(df_gen.Sex == 'M') ].groupby(['Year']).agg({'Height':['mean','std']}).reset_index()
df_gen_female_1 = df_gen[(df_gen.Sex == 'F')].groupby(['Year']).agg({'Height':['mean','std']}).reset_index()
```

```python
# print('.. male .. \n',len(df_gen_male))
# print('.. female .. \n',len(df_gen_female))
# print(df_gen_female_1.Height['mean'],'\n',df_gen_female_1.Height['std'])
# print(df_gen_male_1.Height['mean'],'\n',df_gen_male_1.Height['std'])
```

```python
print('QUESTION 2')
fig, ax = plt.subplots(figsize=(8,6))

avg2 = df_gen_female_1.Height['mean']
std2 = df_gen_female_1.Height['std']
avg1 = df_gen_male_1.Height['mean']
std1 = df_gen_male_1.Height['std']

f1 = avg2 + std2
f2 = avg2 - std2
m1= avg1 + std1
m2 = avg1 - std1

ax.plot(df_gen_male['Year'],df_gen_male['Height'],c="blue")
ax.plot(df_gen_female['Year'],df_gen_female['Height'],c="red")

ax.fill_between(df_gen_female['Year'],f1,f2, color='red', alpha=0.1)
ax.fill_between(df_gen_male['Year'],m1,m2, color='blue', alpha=0.1)

ax.set_yticks([140,150,160,170,180,190,200])
ax.set_yticklabels(['140','150','160','170','180','190','200'])
ax.set_ylim(140,200)
ax.set_xticks([1900,1920,1940,1960,1980,2000,2020])
ax.set_xticklabels(['1900','1920','1940','1960','1980','2000','2020'])
ax.set_xlim(1890,2030)
ax.set_xlabel('Year')
ax.set_ylabel('Height (cm)')
ax.set_title('Average Olympian Height (+/- std dev) in the Summer Olympics')
ax.legend(loc="lower left",framealpha=1, frameon=True, labels=['Male','Female'])
```

QUESTION NO 3

The bar plot below shows the top 10 most decorated Olympians. That is, the Olympic athletes that
have won the highest total number of medals. Each bar is colour coded according to the number of
gold, silver and bronze medals won by that athlete. In this way, the total number of medals won is
represented, along with the breakdown per medal type.Recreate the above bar plot using Matplotlib. The colours for the medal types are #FFD700,#C0C0C0, and #B08D57. Make sure to include the axes labels and plot title. The bars should be
ordered as given in the figure.

```python
df_med = df[['Name','Medal']].dropna()
df_med = df_med.groupby(['Name','Medal']).Medal.count().reset_index(name='Medal Count')
df_med
```

```python
df_med_1 = df_med.pivot(index='Name',columns="Medal",values="Medal Count")
df_med_1['total'] = df_med_1.sum(axis=1)
df_med_1=df_med_1.sort_values(by=['total','Gold','Silver','Bronze'],ascending=False).reset_index()
df_med_1=df_med_1.head(10)
df_med_1
```

```python
print('QUESTION 3')
plt.figure(figsize=(8, 6))
plt.barh(df_med_1['Name'],df_med_1['Gold'],color='#FFD700')
plt.barh(df_med_1['Name'],df_med_1['Silver'],left=df_med_1['Gold'],color='#C0C0C0')
plt.barh(df_med_1['Name'],df_med_1['Bronze'],left=df_med_1['Gold']+df_med_1['Silver'],color='#B08D57')
plt.gca().invert_yaxis()
plt.title('The Most Decorated Olympians of All Time')
plt.ylabel('Name')
plt.xlabel('Medal count')
```
