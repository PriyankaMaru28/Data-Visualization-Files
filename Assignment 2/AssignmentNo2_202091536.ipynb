QUESTION NO 1

```python
df_population=df[df.Country=="Canada"]
```

```python
df_population=df_population[(df_population.Year >= 1955)].sort_values(by='Year')
```

```python
print("Q1. Recreate the following plot using Matplotlib’s functional API. This line plot shows the population of Canada between the years 1955 and 2020.\n")
x=df_population['Year']
y=df_population['Population']
fig, ax = plt.subplots(1)
ax.set_yticks([15e6,20e6,25e6,30e6,35e6])
ax.set_yticklabels(['15M','20M','25M','30M','35M'])
ax.set_title("Canada's Population Growth")
ax.set_xlabel('Year')
ax.set_ylabel('Population')
ax.set_ylim((14.5e6,40e6))
ax.plot(x,y)
```

QUESTION NO 2

```python
df_urban=df[(df.Year <= 2020) & (df.Year >= 1955)]
```

```python
df_urban=df_urban.groupby(["Year"]).sum().reset_index()
df_urban
```

```python
df_urban['PercentageIncrease']= df_urban['UrbanPopulation']/df_urban['Population'] * 100
df_urban.sort_values(by="Year")
df_urban
```

```python
print("Q2. Recreate the following plot using Matplotlib’s functional API. This line plot shows the percentage of the world’s population that are living in urban areas. This has increased from low 30% to mid 50% over the past 60 years.\n")
x=df_urban['Year']
y=df_urban['PercentageIncrease']
plt.yticks(ticks=[30,35,40,45,50,55],labels=['30%','35%','40%','45%','50%','55%'])
plt.ylim((30,58))
plt.title('Percentage of Global Population living in Urban Areas')
plt.xlabel('Year')
plt.ylabel('Urban Percentage')
plt.plot(x, y)
```

QUESTION NO 3

```python
df_abs=df.groupby(by=["Continent","Year"]).sum().reset_index()
df_abs
```

```python
df_abs_africa = df_abs[df_abs.Continent == "Africa"]
df_abs_asia = df_abs[df_abs.Continent == "Asia"]
df_abs_euro = df_abs[df_abs.Continent == "Europe"]
df_abs_latamcar = df_abs[df_abs.Continent == "Latin America and the Caribbean"]
df_abs_na = df_abs[df_abs.Continent == "North America"]
df_abs_oceania = df_abs[df_abs.Continent == "Oceania"]
```

```python
print("Q3. Recreate the following plot using Matplotlib’s object-oriented API. This line plot shows the yearly population change (absolute number) per continent over the past 60 years. You will need to compute this data on a per continent basis using the country values. Make sure to include the axes labels and plot title, and that the colours of each line match the figure.")
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(df_abs_africa['Year'],df_abs_africa['YearlyChange'])
ax.plot(df_abs_asia['Year'],df_abs_asia['YearlyChange'])
ax.plot(df_abs_euro['Year'],df_abs_euro['YearlyChange'])
ax.plot(df_abs_latamcar['Year'],df_abs_latamcar['YearlyChange'])
ax.plot(df_abs_na['Year'],df_abs_na['YearlyChange'])
ax.plot(df_abs_oceania['Year'],df_abs_oceania['YearlyChange'])

ax.legend(loc="center",framealpha=1, frameon=False, labels=['Africa','Asia','Europe','Latin America and the Caribbean','North America','Oceania'])
ax.set_xlabel('Year')
ax.set_ylabel('Yearly Population Change')
ax.set_title('Yearly Population Change by Continent')
```

QUESTION NO 4

```python
df_age = df[(df.Year > 2019)]
df_age
```

```python
df_age = df_age.drop_duplicates().groupby(["Continent","Country","Year"]).mean().reset_index().sort_values(by="Country")
df_age
```

```python
len(df_age)
```

```python
df_age_africa = df_age[df_age.Continent == "Africa"]
df_age_asia = df_age[df_age.Continent == "Asia"]
df_age_euro = df_age[df_age.Continent == "Europe"]
df_age_latamcar = df_age[df_age.Continent == "Latin America and the Caribbean"]
df_age_na = df_age[df_age.Continent == "North America"]
df_age_oceania = df_age[df_age.Continent == "Oceania"]
```

```python
print("Q4. Recreate the following plot using Matplotlib’s object-oriented API. This scatter plot shows the median age and percentage population growth (Yearly%Change) from the year 2019 to 2020 for all the countries of the world. Each country is colour coded according to its continent.")
fig, ax = plt.subplots(figsize=(7.5,5.5))

ax.scatter(df_age_africa['MedianAge'],df_age_africa['Yearly%Change'])
ax.scatter(df_age_asia['MedianAge'],df_age_asia['Yearly%Change'])
ax.scatter(df_age_euro['MedianAge'],df_age_euro['Yearly%Change'])
ax.scatter(df_age_latamcar['MedianAge'],df_age_latamcar['Yearly%Change'])
ax.scatter(df_age_na['MedianAge'],df_age_na['Yearly%Change'])
ax.scatter(df_age_oceania['MedianAge'],df_age_oceania['Yearly%Change'])
ax.set_xticks([15,20,25,30,35,40,45,50])
ax.set_xticklabels([15,20,25,30,35,40,45,50])
ax.set_xlim((14,50))
ax.set_yticks([-2,-1,0,1,2,3,4])
ax.set_yticklabels(['-2','-1','0','1','2','3','4'])
ax.set_ylim((-2.7,4.2))
ax.legend(loc="lower left",framealpha=1, frameon=False, labels=['Africa','Asia','Europe','Latin America and the Caribbean','North America','Oceania'])
ax.set_xlabel('Median Age')
ax.set_ylabel('Percent')
ax.set_title('Country Population Growths (2019 to 2020)')
```
