```python
df_abs=df_pop.groupby(by=["Continent","Year"]).sum().reset_index()
df_abs
```

```python
df_abs_africa = df_abs[df_abs.Continent == "Africa"]
df_abs_asia = df_abs[df_abs.Continent == "Asia"]
df_abs_euro = df_abs[df_abs.Continent == "Europe"]
df_abs_latamcar = df_abs[df_abs.Continent == "Latin America and the Caribbean"]
df_abs_na = df_abs[df_abs.Continent == "North America"]
df_abs_oceania = df_abs[df_abs.Continent == "Oceania"]
df_abs_africa
```

```python
df_olym_count = df_olym[['ID','Year','Team']][(df_olym.Year == 2012) | (df_olym.Year == 2016)]

con_list = ['Great Britain', 'United States', 'Russia', 'Australia','Germany', 'China', 'France', 'Japan', 'Italy', 'Spain', 'Brazil']
df_olym_count = df_olym_count[df_olym_count['Team'].isin(con_list)]
df_olym_count = df_olym_count.groupby(['ID','Year','Team']).count().reset_index().sort_values(by=['ID'],ascending=True)

#df_olym_2012 = df_olym_2012.drop(index=[38,198])
#df_olym_2012 = df_olym_2012.head(11)
print(len(df_olym_count))
df_olym_count
```

```python
df_olym_count_1 = df_olym_count.groupby(['ID','Year','Team']).ID.count().reset_index(name="Count")
df_olym_count_1 = df_olym_count_1.pivot(index=['ID','Team'],columns="Year",values="Count").reset_index()
df_olym_count_2 = df_olym_count_1.dropna()
df_olym_count_1 = df_olym_count_1.fillna(0)
df_olym_count_1
df_olym_count_2 
```

```python
df_olym_count_2012 = df_olym_count_1[['Team',2012]].groupby(['Team']).sum().reset_index().sort_values(by=[2012],ascending= False)
df_olym_count_2012
```

```python
df_olym_count_2016 = df_olym_count_1[['Team',2016]].groupby(['Team']).sum().reset_index().sort_values(by=[2016],ascending= False)
df_olym_count_2016
```

```python
df_olym_count_2 = df_olym_count_2.groupby(['Team']).sum().reset_index()
df_olym_count_2 = df_olym_count_2.rename({2012:'Both'}, axis="columns")
df_olym_count_3 = df_olym_count_2[['Team','Both']]
df_olym_count_3
```

```python
df_country = pd.merge(df_olym_count_2012,df_olym_count_2016,on="Team",how="outer").sort_values(by=[2012], ascending=False)

df_country_3 = df_country.merge(df_olym_count_3,on='Team',how="outer")
df_country
```

```python
df_country_3 
```

```python
df_country = df_country.melt(id_vars="Team", var_name="Year")
df_country
```

```python
print(df_country_3[2012].sum())
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
# Set the figure size
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.barplot(x="value", y="Team", hue="Year", data=df_country, palette=('tab20'))

ax.set(ylabel="", xlabel="Number of Athletes")
ax.set_title('Change in Number of Athletes per Country from 2012 to 2016')
```

```python
# Venn diagram
venn = venn2(subsets=(df_country_3[2012].sum() - df_country_3['Both'].sum(),df_country_3[2016].sum() - df_country_3['Both'].sum(),df_country_3['Both'].sum()), set_labels=('2012','2016'))

ax.legend(handles=[venn],labels=['Venn Diagram'])
```
