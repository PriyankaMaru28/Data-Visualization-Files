```python
# Create the dropdown widget
dropdown = dropdown = widgets.Dropdown(
    options=[('Natural', 1), ('Lab', 2), ('All', 3)],
    value=3,
    description='Origin:',
)

# Display the dropdown widget
display(dropdown)

def dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        clear_output(wait=True)
        display(dropdown)
        if change['new'] == 2 :
            df_diamond_11 = df_diamond.loc[df_diamond.type == 'lab']
        elif change['new'] == 1 :
             df_diamond_11 = df_diamond.loc[df_diamond.type == 'natural']
        else :
            df_diamond_11 = df_diamond
        plotting(df_diamond_11)
   
def plotting(df):
        df_diamond_12 = df.groupby(['colour','clarity']).mean().reset_index().sort_values(by=['colour','clarity'],ascending=False)
        df_diamond_12 = df_diamond_12.pivot(index='colour', columns='clarity', values='carat')
        df_diamond_12 = df_diamond_12.reindex(columns=['SI2', 'SI1', 'VS2','VS1','VVS2','VVS1','IF'])
        plt.figure(figsize=(10,8))
        sns.heatmap(data=df_diamond_12, cmap='mako')
  
dropdown.observe(dropdown_change)
```

```python
# Create text input widget
names = widgets.Text(placeholder='Enter Baby Names')

# Create label widget
label = widgets.Label(value='Enter Names:')

# Create horizontal box layout widget for label and input
input_box = widgets.HBox([label, names])

entered_names = []
#Define function to process input
def input_change(change):
    names = change['new'].split(' ') # Split input string by space
    names = [name.upper() for name in names] # Remove leading/trailing spaces
    global entered_names
    entered_names = names
    print(entered_names)
    

# Observe input changes and call processing function
names.observe(input_change, names='value')

toggle = widgets.ToggleButtons(
    options=['Baby boy names','Baby girl names'],
    disabled=False,
    button_style='', 
    layout=widgets.Layout(width='1000px')
)


# Create vertical box layout widget for input box
widget_box = widgets.VBox([input_box])

# Display widgets
display(widget_box)
display(toggle)
# plt.clf()
# plt.cla()
# plt.close()

plt.figure(figsize=(25,23))
#color array
color = ['blue','green','violet', 'orange', 'pink', 'red']
df_new=pd.DataFrame()
def on_changes(change):
    if change['type'] == 'change' and change['name'] == 'value':
        clear_output(wait=True)
        display(widget_box)
        display(toggle)
        if change['new'] == "Baby girl names" :
            df_new = girl_diction
        else :
            df_new = boy_diction
            print("Boy")
        keysList = list(df_new.keys())
        i=0
        for x in keysList:
          if x in entered_names:
              if i <=5:
                plt.plot(df_new[x]['year'], df_new[x]['rank'], color=color[i], zorder=10, linestyle='-', marker='o')
                plt.legend(labels=entered_names,loc="lower right")
                i=i+1
              else:
                plt.plot(df_new[x]['year'], df_new[x]['rank'], color='silver')
          else:
            plt.plot(df_new[x]['year'], df_new[x]['rank'], color='silver')
        plt.gca().invert_yaxis()
        plt.xticks([1996,1998,2000,2002,2004,2006,2008,2010,2012,2014,2016,2018,2020])
        plt.yticks([1, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]) 
        plt.ylim((1100, 1)) 
        plt.xlabel('Year')
        plt.ylabel('Rank')
        plt.grid(axis='y')
        sns.despine(left=True)
        plt.show()

toggle.observe(on_changes)
```
